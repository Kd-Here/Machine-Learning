```python
# fig,(ax) = plt.subplot(a,b)
# Here fig represent the ploting fig
# ax are the number of plotting figures for subplots along x axis how much variables if you want to plot fig in ax give (ax1,ax2)
# To obtain side-by-side subplots, pass parameters 1, 2 for one row and two columns.


from numpy import random
u = random.randint(100, size=(12))
print(u)

v = random.randint(200,size=(12))
print(v)

# always rememeber the subplots not subplot
fig, (ax1, ax2) = plt.subplots(1,2)
fig.suptitle('Horizontally stacked subplots')
ax1.scatter(u,v,marker="*",color='r')
# We are scattering points on first fig
ax1.plot(u+v,v,color='y') # we are plotting on same fig so use same name i.e ax1

ax2.scatter(u*2,v*4,marker="+",color='y')
ax2.scatter(u+2*v,v+2*u,marker="+",color='blue')
```

```python
fig, axs = plt.subplots(2, 2)

# When stacking in two directions, the returned axs is a 2D NumPy array.

axs[0, 0].scatter(u, v,color='blue')
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].scatter(u*2, v*2,color='green')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].scatter(u*3,v*3,color='yellow')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].scatter(u*5, v*5,color='red')
axs[1, 1].set_title('Axis [1, 1]')
```
